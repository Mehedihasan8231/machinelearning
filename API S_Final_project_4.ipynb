```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
import pandas as pd

df = pd.read_csv('/content/drive/MyDrive/DM&KD/Market_Basket_Optimisation.csv')

#Check number of rows and columns in the dataset
print("The dataset has %d rows and %d columns." % df.shape)
```

```python
X = df.iloc[:, [3, 4]].values
```

```python
from sklearn.cluster import KMeans
# (Within-Cluster Sum of Square)
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
import scipy.cluster.hierarchy as sch

plt.figure(figsize=(10,10))
dendrogram = sch.dendrogram(sch.linkage(X, method = 'complete'))
plt.title('Dendrogram')
plt.xlabel('source')
plt.ylabel('Euclidean distances')
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering

hc = AgglomerativeClustering(n_clusters = 3, affinity = 'euclidean', linkage = 'single')

y_hc = hc.fit_predict(X)
```

```python
# Visualising the clusters

plt.figure(figsize=(8,8))
plt.scatter(X[y_hc == 0, 0], X[y_hc == 0, 1], s = 100, c = 'red', label = 'short time ueser')
plt.scatter(X[y_hc == 1, 0], X[y_hc == 1, 1], s = 100, c = 'blue', label = 'mid time ueser')
plt.scatter(X[y_hc == 2, 0], X[y_hc == 2, 1], s = 100, c = 'green', label = 'high time ueser')
plt.title('Clusters of user using Hierarchical Clustering')
plt.xlabel('inter_api_access_duration(sec)')
plt.ylabel('api_access_uniqueness')
plt.legend()
plt.show()
```

k-means Clustering

```python
kmeans = KMeans(n_clusters = 3, init = 'k-means++', random_state = 42)
y_kmeans = kmeans.fit_predict(X)
```

```python
plt.figure(figsize=(8,8))
plt.scatter(X[y_kmeans == 0, 0], X[y_kmeans == 0, 1], s = 100, c = 'red', label = 'active ueser')
plt.scatter(X[y_kmeans == 1, 0], X[y_kmeans == 1, 1], s = 100, c = 'blue', label = 'mid time ueser')
plt.scatter(X[y_kmeans == 2, 0], X[y_kmeans == 2, 1], s = 100, c = 'green', label = 'short time ueser')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 100, c = 'black')
plt.title('Clusters of Customers using K-Means Clustering')
plt.xlabel('sequence_length(count)')
plt.ylabel('inter_api_access_duration(sec)')
plt.legend()
plt.show()
```
